```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Creiamo un array di dati con NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creiamo un grafico con Matplotlib
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Grafico di una Funzione Sinusoidale")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")
plt.grid(True)
plt.show()

# Stampa le versioni delle librerie per verifica
print(f"Versione di NumPy: {np.__version__}")
print(f"Versione di Matplotlib: {matplotlib.__version__}")
```
